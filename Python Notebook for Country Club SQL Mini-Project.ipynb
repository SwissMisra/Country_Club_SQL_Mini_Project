#### Q1: Some of the facilities charge a fee to members, but some do not. Write a SQL query to produce a list of the names of the facilities that do. ####

#### Q2: How many facilities do not charge a fee to members? ####

```python
%%sql
SELECT COUNT( `facid` ) 
FROM `Facilities` 
WHERE `membercost` = 0
```

#### Q3: Write an SQL query to show a list of facilities that charge a fee to members, where the fee is less than 20% of the facility's monthly maintenance cost. Return the facid, facility name, member cost, and monthly maintenance of the facilities in question. ####

```python
%%sql
SELECT `facid` , `name` , `membercost` , `monthlymaintenance` 
FROM `Facilities` 
WHERE `membercost` * .2 < `monthlymaintenance`
```

#### Q5: Produce a list of facilities, with each labelled as 'cheap' or 'expensive', depending on if their monthly maintenance cost is more than $100. Return the name and monthly maintenance of the facilities n question. ####

```python
%%sql
SELECT `name` , `monthlymaintenance`, 
CASE 
    WHEN `monthlymaintenance` > 100 
        THEN 'expensive' 
    WHEN `monthlymaintenance` < 100 
        THEN 'cheap' 
    END AS type 
FROM `Facilities`
    
```

#### Q7: Produce a list of all members who have used a tennis court. Include in your output the name of the court, and the name of the member formatted as a single column. Ensure no duplicate data, and order by the member name. ####

```python
%%sql
SELECT DISTINCT firstname||' '||m.surname as name
    FROM Members as m
INNER JOIN Bookings as b
USING(memid)
WHERE b.facid in (0,1)
```

#### Q8: Produce a list of bookings on the day of 2012-09-14 which will cost the member (or guest) more than $30. Remember that guests have different costs to members (the listed costs are per half-hour 'slot'), and the guest user's ID is always 0. Include in your output the name of the facility, the name of the member formatted as a single column, and the cost. Order by descending cost, and do not use any subqueries. ####

```python
%%sql
SELECT f.name, m.firstname||' '||m.surname as member_name, f.membercost, f.guestcost, b.starttime
FROM Members as m
LEFT JOIN Bookings as b
USING (memid)
LEFT JOIN Facilities as f
ON f.facid = b.facid
WHERE membercost > 30 OR guestcost > 30 AND starttime LIKE '%09-14%'
ORDER BY f.membercost, f.guestcost DESC
```

#### Q9: This time, produce the same result as in Q8, but using a subquery. ####

```python
%%sql
SELECT DISTINCT member_name, sub.name, sub.cost, sub.starttime
FROM (
    SELECT 
    m.firstname||' '||m.surname as member_name, 
    b.starttime,
    f.name,
    CASE
        WHEN membercost > 30 THEN membercost
        WHEN guestcost > 30 THEN guestcost
        END as cost
    FROM Members as m
    LEFT JOIN Bookings as b
    ON b.memid = m.memid
    LEFT JOIN Facilities as f
    ON f.facid = b.facid) as sub
    WHERE starttime LIKE '%09-14%' AND sub.cost IS NOT NULL
```

#### Q10: Produce a list of facilities with a total revenue less than 1000. The output of facility name and total revenue, sorted by revenue. Remember that there's a different cost for guests and members! ####

```python
%%sql
SELECT f.name,
        (f.guestcost * COUNT(CASE WHEN b.memid=0 THEN 1 END) +
        f.membercost * COUNT(CASE WHEN b.memid!=0 THEN 1 END))
        AS revenue
    FROM Facilities as f
    LEFT JOIN Bookings as b
    ON b.facid = f.facid
    GROUP BY f.facid
    ORDER BY revenue DESC
```

#### Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order ####

```python
%%sql
SELECT m.surname||', '||m.firstname as name, m2.firstname||' '||m2.surname AS recommender
FROM Members as m
LEFT JOIN Members as m2
ON m.recommendedby = m2.memid
ORDER BY name
```

#### Q12: Find the facilities with their usage by member, but not guests ####

```python
%%sql
SELECT sub.name, 
sub.Tennis_Court_1, 
sub.Tennis_Court_2, 
sub.Badminton_Court, 
sub.Table_Tennis, 
sub.Massage_Room_1, 
sub.Massage_Room_2, 
sub.Squash_Court, 
sub.Snooker_Table,
sub.Pool_Table
FROM (
    SELECT m.surname||' '||m.firstname as name,
    COUNT(CASE WHEN b.memid = m.memid AND f.facid=0 THEN 1 END) AS Tennis_Court_1,
    COUNT(CASE WHEN b.memid = m.memid AND f.facid=1 THEN 1 END) AS Tennis_Court_2,
    COUNT(CASE WHEN b.memid = m.memid AND f.facid=2 THEN 1 END) AS Badminton_Court,
    COUNT(CASE WHEN b.memid = m.memid AND f.facid=3 THEN 1 END) AS Table_Tennis,
    COUNT(CASE WHEN b.memid = m.memid AND f.facid=4 THEN 1 END) AS Massage_Room_1,
    COUNT(CASE WHEN b.memid = m.memid AND f.facid=5 THEN 1 END) AS Massage_Room_2,
    COUNT(CASE WHEN b.memid = m.memid AND f.facid=6 THEN 1 END) AS Squash_Court,
    COUNT(CASE WHEN b.memid = m.memid AND f.facid=7 THEN 1 END) AS Snooker_Table,
    COUNT(CASE WHEN b.memid = m.memid AND f.facid=8 THEN 1 END) AS Pool_Table
    FROM Members as m
    LEFT JOIN Bookings as b
    ON b.memid = m.memid
    LEFT JOIN Facilities as f
    ON f.facid = b.facid
    WHERE m.memid!=0
    GROUP BY m.memid) as sub
```

#### Q13: Find the facilities usage by month, but not guests ####

```python
%%sql 
SELECT sub.name, sub.July_usage, sub.August_usage, sub.Sept_usage
FROM
(SELECT 
COUNT(CASE WHEN b.memid!=0 AND b.starttime LIKE '%2012-07%' THEN 1 END) as July_usage,
COUNT(CASE WHEN b.memid!=0 AND b.starttime LIKE '%2012-08%' THEN 1 END) as August_usage,
COUNT(CASE WHEN b.memid!=0 AND b.starttime LIKE '%2012-09%' THEN 1 END) as Sept_usage,
f.name
FROM Bookings as b
LEFT JOIN Facilities as f
ON f.facid = b.facid
GROUP BY f.name) as sub
```
